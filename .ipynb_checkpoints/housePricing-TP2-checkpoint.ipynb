```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# loading the data
df = pd.read_csv('datasets/housing.csv')
df.head()
```

```python
df['ocean_proximity'].value_counts()
```

```python
ocean = pd.get_dummies(df.ocean_proximity)
ocean.columns = ['ocean_1','ocean_2','ocean_3','ocean_4','ocean_5']
df = df.drop("ocean_proximity",axis = 1)
full_df = pd.concat([df,ocean],axis = 1)
```

```python
full_df[['ocean_1','ocean_2','ocean_3','ocean_4','ocean_5']] = full_df[['ocean_1','ocean_2','ocean_3','ocean_4','ocean_5']].astype(float)
```

```python
full_df.head()
```

### here we splitted our data to train data and test data

```python
train_df = full_df[:16512]
test_df = full_df[16512:]
```

#### since total_bedrooms has  a weak correlation with the output variable it is better to delete it
#### we'll delete also some other non correlated features

```python
full_df = full_df.drop(['ocean_1','ocean_2','ocean_3','ocean_4','ocean_5','total_bedrooms'], axis = 1)
```

```python
full_df.columns
```

```python
train_df = full_df[:16512]
test_df = full_df[16512:]
```

```python
test_df.shape
```

```python
# extraction of X_train (n_samples, n_features) and y_train (target variable)
X_train = train_df.drop("median_house_value", axis=1) 
y_train = train_df["median_house_value"].to_numpy()
print('X_train:', X_train.shape, '; y_train:', np.shape(y_train))
```

```python
# loading the training data
# extraction of X_test and y_test
X_test= test_df.drop("median_house_value", axis=1) 
y_test = test_df["median_house_value"].to_numpy() 
print('X_test:', X_test.shape, '; y_test:', np.shape(y_test))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
# here we can see how many paremters we can deal with iin each algorithm
print(LinearRegression().get_params())
print(DecisionTreeRegressor().get_params())
print(RandomForestRegressor().get_params())
```

```python
# K-fold cross-validation and GridSearchCV
pipelines = []
params = []
names = []

#
# ajouter LinearRegression
pipelines.append(Pipeline([('clf', LinearRegression())])) ### LinearRegression
params.append({'clf__normalize':[True]})
names.append('LinearRegression')

# ajouter DecisionTreeRegressor
pipelines.append(Pipeline([('clf', DecisionTreeRegressor())])) ## DecisionTreeRegressor
params.append({'clf__max_depth':np.linspace(5, 15, 5)})
names.append('DecisionTreeRegressor') 


pipelines.append(Pipeline([('clf', RandomForestRegressor())])) ## RandomForestRegressor
params.append({'clf__n_estimators': [100,200]})
names.append('RandomForestRegressor')
```

```python
from sklearn.model_selection import KFold, GridSearchCV, cross_val_score

def model(pipeline, parameters, name, X, y):    
    cv = KFold(n_splits=5, shuffle=True, random_state=32)
    grid_obj = GridSearchCV(estimator=pipeline, param_grid=parameters, cv=cv, scoring='r2', n_jobs=-1)
    grid_obj.fit(X,y)  
    print(name, 'R2:', grid_obj.best_score_)
    print(name, 'R2:', grid_obj.best_params_)
    estimator = grid_obj.best_estimator_
    estimator.fit(X,y) # training sur tout training dataset
    return estimator 
estimators = []
for i in range(len(pipelines)):    
    estimators.append(model(pipelines[i], params[i], names[i], X_train, y_train))
```

```python
from sklearn.metrics import mean_squared_error, r2_score 

# extraction of X_test and y_test
X_test= test_df.drop("median_house_value", axis=1) 
y_test = test_df["median_house_value"].to_numpy() 
print('X_test:', X_test.shape, '; y_test:', np.shape(y_test))

# Evaluation

for i, estimator in enumerate(estimators):
    print('\nPerformance :', names[i])
    y_pred = estimator.predict(X_test)
    print('\n mean_squared_error :', mean_squared_error(y_test, y_pred))    
    print('\n r2_score :', r2_score(y_test, y_pred))
```

### in the cell below we'll save each of our models for  future uses 

```python
import pickle
# save the classifier
for i, estimator in enumerate(estimators):
    with open(names[i]+".pkl", 'wb') as fid:
        pickle.dump(estimator, fid) 
```
